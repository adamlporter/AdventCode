Passports have key:value pairs for eight keys. Which ones have all eight fields?

```python
import re
```

```python
test_data = '''ecl:gry pid:860033327 eyr:2020 hcl:#fffffd
byr:1937 iyr:2017 cid:147 hgt:183cm

iyr:2013 ecl:amb cid:350 eyr:2023 pid:028048884
hcl:#cfa07d byr:1929

hcl:#ae17e1 iyr:2013
eyr:2024
ecl:brn pid:760753108 byr:1931
hgt:179cm

hcl:#cfa07d eyr:2025 pid:166559648
iyr:2011 ecl:brn hgt:59in'''

data = [x for x in test_data.split('\n\n')]

reg_exp = re.compile("([a-z]{3}):(\S*)")

tally = 0

for chunk in data:
    input = ' '.join(chunk.split('\n')) # get rid of \n characters in input
    key_vals = re.findall(reg_exp,input) # extract key, value from string
    dd = {}
    for tup in key_vals:
        dd[tup[0]] = tup[1]
    if len(dd) == 8:
        tally += 1
    elif len(dd) == 7 and 'cid' not in dd:
        tally += 1

print(tally)

```
